```python
import pandas as pd
import numpy as np
import datetime
import math
game_data =pd.io.pickle.read_pickle('/Users/theblooms/Desktop/insight/data/4403425-14-09-09-11-42.pkl')
```

```python
sh = game_data.section.str.extract('(?P<letter>[A-Z]*)(?P<digit>\d*)')
game_data= game_data.join(sh)
game_data= game_data[game_data.digit != ""]
```

```python
game_data['digit'] = game_data['digit'].apply(int)
game_data['price'] = game_data['price'].apply(float)
game_data['deck']= np.floor_divide(game_data['digit'],100)
game_data['block']= game_data['digit']-game_data['deck']*100
game_data.head()
```

```python
a=game_data.pivot_table(['price'],index=['deck'],columns=['block'])



```

```python
def heatmap(df, cmap=plt.cm.RdYlBu):
 fig = plt.figure()
 ax = fig.add_subplot(111)
 axim = ax.imshow(df.values, cmap=cmap, interpolation='nearest')
 ax.set_xlabel(df.columns.name)
 ax.set_xticks(np.arange(len(df.columns)))
 ax.set_xticklabels(list(df.columns))
 ax.set_ylabel(df.index.name)
 ax.set_yticks(np.arange(len(df.index)))
 ax.set_yticklabels(list(df.index))
 ax.set_aspect(5./1.)
 plt.colorbar(axim)

heatmap(a,)
```
